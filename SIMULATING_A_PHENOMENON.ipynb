### Regression

In this part of my project I will talk about rgression analysis. Regression analysis is a phenomenon of interest and have a number of observations. Following the assumption that (at least) one of the features depends on the others, you try to establish a relation among them. 

"Reg Plot" and "LM Plot" are two important types of figure that we can plot using Seaborn. They have similar appearance, but have different functionality.

***Regression Line*** is a straight line that describes how a response variable y changes as an explanatory variable x changes.

```python
sns.set(color_codes=True)
sns.lmplot(x = "Grade", y = "Hours", data=project_2019);  # function that is used to visualize a linear relationship as determined through regression.
```

The Linear Line is the best trend of the Grade in accordance of the Hours studied. The points are hours that the students studying per week and the points that are far away from this line are called outliers in the dataset. The algorithm predict the best fit for understanding the trend in behavior of Grade. This shadow is actually the statistical mean that means the prediction of Hours in this module. We can use all information to better understand the data adding more features to the plot and for this we will use Hue parameter.

```python
sns.lmplot(x="Grade", y="Hours", hue="Logins", data=project_2019); # function divide the data per column and show the variables.
```

Right now, the same plot on separate facets. I used the parameter col the split the data: 

```python
sns.lmplot(x = "Grade", y = "Hours", col = "Logins", data=project_2019);
```

Let us visualize one more on built-in *project_2019*:

```python
sns.lmplot(x = "Grade", y = "Hours", row = "Qualification", data=project_2019);
```

Plot of nonparametric regression using a lowess smoother. It is a tool that creates a smooth line to help to see relationship between variables.

```python
sns.lmplot(x="Grade", y="Hours", data=project_2019, lowess=True);
```

This graphic performs a simple linear regression model fit and plot.
If you compare the lmplot plot and regplot plot you will see that the resulting are identical differentiating only the shape.

```python
sns.regplot(x="Grade", y="Hours", data=project_2019, color="y" ); # function that is used to visualize a linear relationship as determined through regression.
```

Controlling the size and shape of the plot.

```python
f, ax = plt.subplots(figsize=(10, 6)) # the fig size can be changeble as requerid.
sns.regplot(x="Grade", y="Hours", data=project_2019, ax=ax);
```

**lmplot** is controlled through the FacetGrid interface using the size and aspect parameters, which apply to each facet in the plot, not to the overall figure itself.

```python
sns.lmplot(x="Grade", y="Hours", col="Qualification", data=project_2019, col_wrap=2, height=3);
```

##### Plotting a regression in other contexts.

More complex plot that uses jointplot with regplot to show the linear regression.

```python
sns.jointplot(x="Grade", y="Hours", data=project_2019, kind="reg");
```

I created the graphics of predictions with a Linear Regression's model.
I defined new **X** and **Y**.

```python
X = project_2019[['Grade']]
Y = project_2019[['Hours']]
```

I need to import train test and linear regression form ***sklearn*** library. 

```python
from sklearn.model_selection import train_test_split  # import scilkit-learn 
from sklearn.linear_model import LinearRegression
```

I splitted the data in to data test and data train:

```python
X_train, X_test , y_train , y_test = train_test_split(X,Y,test_size=0.25,random_state=26)
```

```python
model = LinearRegression() # define X_train and Y_train
model.fit(X_train, y_train)
```

```python
predictions=model.predict(X_test) # run predict
```

```python
sns.distplot(y_test-predictions);
```
